## Defining Filters

```python
from PIL import Image
def padding(image, num):
  pil_image=Image.fromarray(image)
  width, height = pil_image.size
  new_width = width + right + left
  new_height = height + top + bottom
  result = Image.new(pil_image.mode, (new_width, new_height), (0))
  result.paste(pil_image, (left, top))
  np_array = np.array(result)
  return np_array
```

```python
for i in images:
  num = 1
  padded = padding(i, num)
  print(i.shape)
  print(padded.shape)
  plt.imshow(padded)
  plt.show()
```

```python
def filters(image):
  filter = np.array ([1,1,1,0,0,0,-1,-1,-1])
  filter = filter.reshape(3,3)
  width , height = image.shape
  z = 3
  filtered = []
  for i in range(0,height - 2):
    z2 = 3
    for j in range(0,width-2):
      z1 = image[i:z, j:z2]
      product = np.multiply(z1,filter)
      sumation = product.sum()
      filtered.append(sumation)
      z2+=1
    z+= 1
  return filtered
```

```python
for i in images:
  filtering = filters(i)
  filter_array = np.array(filtering)
  filtered_image = filter_array.reshape(26,26)
  plt.imshow(filtered_image)
  plt.show()
```
